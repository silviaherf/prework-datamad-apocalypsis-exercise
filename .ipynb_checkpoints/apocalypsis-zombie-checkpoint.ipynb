Cargas un arsenal de armas y una mochila con cargadores de munición de tamaño variable para algunos tipos de armas. ¿Cuántos zombies puedes matar antes de tener que echar a correr por tu vida? Deberás imprimir el resultado de cada caso que corresponde al número de balas que tienes de las armas que cargas.  

ALGORITMO: Por cada arma diferente que cargas, tienes que comprobar si portas cargadores para ella. Si esto es afirmativo deberás sumar el total de balas que podrás disparar. 

## Caso 0: Ejemplo simple

```python
armas = ['pistola','escopeta']

cargadores = {
    'pistola': [10, 10], 
    'escopeta': [2, 2, 2, 2, 2]
}

# your code
import random
balas_pistola=0
balas_escopeta=0
balas_total=0
i=0
j=0

while(i<len(cargadores['pistola'])):
    arma_escogida=random.choice(armas) 
    if arma_escogida=='pistola' :
        balas_pistola+=cargadores['pistola'][i]
        print("Has cogido una pistola")
        print("Has gastado",balas_pistola,"balas de pistola")
        i+=1
    elif (arma_escogida=='escopeta') & (j<len(cargadores['escopeta'])):
        balas_escopeta+=cargadores['escopeta'][j]
        print("Has cogido una escopeta")
        print("Has gastado",balas_escopeta,"balas de escopeta")
        j+=1
        
while(j<len(cargadores['escopeta'])):
        arma_escogida=random.choice(armas) 
        if (arma_escogida=='pistola') & (i<len(cargadores['pistola'])):
            balas_pistola+=cargadores['pistola'][i]
            print("Has cogido una pistola")
            print("Has gastado",balas_pistola,"balas de pistola")
            i+=1
        elif arma_escogida=='escopeta':
            balas_escopeta+=cargadores['escopeta'][j]
            print("Has cogido una escopeta")
            print("Has gastado",balas_escopeta,"balas de escopeta")
            j+=1
    
balas_total=balas_pistola+balas_escopeta        
print("Has usado", balas_total, "balas")
```

**Output esperado** -> 30
